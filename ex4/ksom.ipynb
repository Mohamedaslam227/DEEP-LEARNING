```python
import torch
```

```python
num_outputs = 2
num_inputs = 4
alpha = 1
epochs = 10
inputs = torch.tensor([[1,0,0,1],[0,0,0,1],[1,0,1,1],[0,1,1,0]])
weights = torch.rand(num_outputs,num_inputs)
```

```python
print(inputs)
print(weights)
```

```python
def squared_euclidean(inputs,weights):
    res = 0
    for i,j in zip(inputs,weights):
        res += (i-j)**2
    return res
```

```python
def update_weights(input,minimum):
    for j in range(num_inputs):
        weights[minimum][j] += alpha * (input[j] - weights[minimum][j])
```

```python
def ksom(inputs,weights,epochs):
    for epoch in range(epochs):
        print("\nepoch",epoch)
        for ip in inputs:
            d = []
            for wt in weights:
                res = squared_euclidean(ip,wt)
                d.append(res)
            minimum = d.index(min(d))
            update_weights(ip,minimum)
            print("D(j) : ",d)
            print("Minimum D(j) : ",minimum)
            print("Weights : ",weights)
```

```python
ksom(inputs,weights,epochs)
```
