```python
import torch

tensor_a = torch.tensor([[1, 2], 
                        [3, 4]])

tensor_b = torch.tensor([[5, 6], 
                        [7, 8]])

transposed_tensor = torch.transpose(tensor_a, 0, 1)
print("1. Transposition:")
print(transposed_tensor)

sqrt_tensor = torch.sqrt(tensor_a.float())
print("\n2. Element-wise square root:")
print(sqrt_tensor)

concatenated_tensor = torch.cat((tensor_a, tensor_b), dim=1)
print("\n3. Concatenation along dimension 1:")
print(concatenated_tensor)

abs_tensor = torch.abs(tensor_a - tensor_b)
print("\n4. Element-wise absolute value:")
print(abs_tensor)

sum_along_dimension = torch.sum(tensor_a, dim=1)
print("\n5. Reduction operation - Sum along dimension 1:\n")
print(sum_along_dimension)

a=torch.tensor([1, 4, 6, 8, 10, 14])
cla=torch.clamp(a,min=5,max=10)
print("\n6. Clamp: ",cla)

b=torch.linspace(start=1,end=10,steps=5)
print("\n7. Creating an Equally spaced tensor : \n",b)

c=torch.eye(n=4,m=5)
print("\n8. Creating a tensor with 1's in leading Diagonal : \n",c)

d=torch.full(size=[2,3,4],fill_value=5)
print("\n9. Creating a tensor with a fill value:\n",d)

e=torch.tensor([[1, 2, 6], [3, 4, 7]])
print("\n10. Taking the element at specified index : \n")
torch.take(e,torch.tensor([2,5]))

f=torch.tensor([[1,2,3],
                [4,5,6]])
print("\n11. Removes a tensor dimension : \n")
torch.unbind(f,dim=1)

g=torch.tensor([[1,2],
                [3,4],
                [5,6]])
h=g.clone()
print("\n12. Creating a deepcopy of a tensor : \n")
print(h)
```
